```python
from qiskit import *
from qiskit.visualization import plot_bloch_multivector
from qiskit.quantum_info import Statevector

import numpy as np
```

```python
def test_qc(theta_x, theta_y):
    qc = QuantumCircuit(1)
    qc.rx(theta_x, 0)
    qc.ry(theta_y, 0)
    statevector = Statevector(qc)
    return statevector
```

```python
theta_x = np.pi/3
theta_y = np.pi/7
plot_bloch_multivector(test_qc(theta_x, theta_y))
```
