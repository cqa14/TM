```python
# using matplotlib
# plot sin and cos
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('Sine and Cosine')

# save this file in .tex format using tikzplotlib
import tikzplotlib
tikzplotlib.save("plot.tex")
```
